```python
def is_ascii(s):
    return all(ord(c) < 128 for c in s)

non_english_df = df[df['Description'].apply(lambda x: (isinstance(x, str) and not is_ascii(x)))]
lst = non_english_df["Description"].tolist()
```

```python
import string

def is_text_english_with_numbers(s, threshold=0.95):
    letters = [c for c in s if c.isalpha()]
    
    if not letters:
        return False
    
    ascii_letters = set(string.ascii_letters)
    count_ascii = sum(1 for c in letters if c in ascii_letters)
    ratio = count_ascii / len(letters)
    return ratio >= threshold

non_english_df_2 = df[df['Description'].apply(lambda x: isinstance(x, str) and not is_text_english_with_numbers(x))]
```
